```python
import os
import pandas as pd
import sqlite3

data_dir = './'
synthetic_HF_TRNS_TRAN = pd.read_csv(os.path.join(data_dir, 'synthetic_HF_TRNS_TRAN.csv'))
df = synthetic_HF_TRNS_TRAN

# SQLite 메모리 데이터베이스 연결
conn = sqlite3.connect(':memory:')
df.to_sql('HF_TRNS_TRAN', conn, index=False, if_exists='replace')  # Ensure the table name matches

# SQL 쿼리 실행
query = '''
SELECT 
	tran_dt
	, COUNT(*) AS dps_cnt
	, tran_tmrg
	, dps_fc_sn
	, dps_ac_sn
	, SUM(tran_amt) AS tran_amt
FROM HF_TRNS_TRAN HF
GROUP BY tran_dt, tran_tmrg, dps_fc_sn, dps_ac_sn
ORDER BY tran_dt, tran_tmrg, dps_fc_sn, dps_ac_sn
'''

result = pd.read_sql_query(query, conn)

# 출력
print(result)

# 연결 종료
conn.close()
```

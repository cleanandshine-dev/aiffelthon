```python
import os
import pandas as pd

pd.set_option('display.max_columns', None)
path = '../../../../../archive/raw/HI-Small_Trans.csv'
df = pd.read_csv(path)
```

```python
df.describe()
```

```python
# 1. 필요한 컬럼 선택
amount_received = df['Amount Received']
amount_paid = df['Amount Paid']

num_samples = total_rows = df.shape[0]

bank_ids_min = df['From Bank'].min()
bank_ids_max = df['From Bank'].max()

# 2. 통계 정보 계산
amount_received_mean =  amount_received.mean() 
amount_received_std =  amount_received.std() 
amount_received_skewness =  amount_received.skew() 

amount_paid_mean =  amount_paid.mean() 
amount_paid_std =  amount_paid.std() 
amount_paid_skewness =  amount_paid.skew() 

# 3. "Is Laundering" 컬럼 통계값 계산
is_laundering_counts = df['Is Laundering'].value_counts(normalize=True)
is_laundering_mean =  is_laundering_counts.get(1, 0)   # 1의 비율 (자금 세탁)
is_laundering_std =  is_laundering_counts.get(0, 0)    # 0의 비율 (비자금 세탁)

# 4. 결과 출력
print(f"num_samples = {num_samples}")

print(f"bank_ids_min = {bank_ids_min}")
print(f"bank_ids_max = {bank_ids_max}")

print(f"amount_received_mean = {amount_received_mean}")
print(f"amount_received_std = {amount_received_std}")
print(f"amount_received_skewness = {amount_received_skewness}")
print(f"amount_paid_mean = {amount_paid_mean}")
print(f"amount_paid_std = {amount_paid_std}")
print(f"amount_paid_skewness = {amount_paid_skewness}")

# 5. "Is Laundering" 통계 출력
print(f"\nis_laundering_0 = {is_laundering_mean * 100}")
print(f"is_laundering_1 = {is_laundering_std * 100}")

# 6. 통화 및 결제 형식 정보 출력
print("currencies = ", sorted(df['Receiving Currency'].unique()))
print("payment_formats = ", sorted(df['Payment Format'].unique()))

```
